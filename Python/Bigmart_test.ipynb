#### Fixing wrong data

We are going to look after two ways here

1) Replace

2) Remove

```python
data3 = data2.copy()
```

```python
data3
```

Replace method

```python
data3.loc[1,'Outlet_Size']= 'Medium'
```

```python
data3
```

Replacing value by rule

Here ensuring item weight is maximum 18

```python
for x in data3.index:
    if data3.loc[x,'Item_Weight'] >18:
        data3.loc[x,'Item_Weight'] = 18
print(data3['Item_Weight'])
```

For Loop: Iterates over each index in the DataFrame.

If Statement: Checks a condition and performs an action if the condition is true.

No then Keyword: Python's if statement does not require a then keyword. The block of code to be executed is indented.

data3.index: Provides the row indices of the DataFrame, allowing iteration over each row.

Remove method

```python
data4 = data2.copy()
```

```python
data4.head()
```

```python
for x in data4.index:
    if data4.loc[x,'Item_Weight'] > 18:
        data4.drop(x,inplace = True)
print(data4['Item_Weight'])
```
